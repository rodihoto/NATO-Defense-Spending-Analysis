## Nato Defense Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Country': ['USA', 'Germany', 'France', 'UK', 'Poland'],
        'Defense_Spending_Billion_EUR': [700, 50, 45, 60, 15]}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.bar(df['Country'], df['Defense_Spending_Billion_EUR'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Defense Spending (Billion EUR)')
plt.title('Defense Spending by Country')
plt.show()
plt.savefig('Defense Spending by Country.png')
```

```python
import seaborn as sns

# Sample data
data = {'Country': ['USA', 'Germany', 'France', 'UK', 'Poland'],
        'GDP_Billion_EUR': [21000, 3800, 2700, 3000, 600],
        'Defense_Spending_Billion_EUR': [700, 50, 45, 60, 15]}

df = pd.DataFrame(data)
df['Defense_Percentage_GDP'] = (df['Defense_Spending_Billion_EUR'] / df['GDP_Billion_EUR']) * 100

plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP_Billion_EUR', y='Defense_Percentage_GDP', data=df, s=100)
plt.xlabel('GDP (Billion EUR)')
plt.ylabel('Defense Spending (% of GDP)')
plt.title('Defense Spending as Percentage of GDP')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Defense spending vs. GDP of key NATO countries
data = {
    'Country': ['USA', 'Germany', 'France', 'UK', 'Poland', 'Canada'],
    'GDP_Billion_EUR': [21000, 3800, 2700, 3000, 600, 2200],
    'Defense_Spending_Billion_EUR': [700, 50, 45, 60, 15, 30]
}

df = pd.DataFrame(data)
df['Defense_Percentage_GDP'] = (df['Defense_Spending_Billion_EUR'] / df['GDP_Billion_EUR']) * 100

# Create a pie chart to visualize the share of defense spending among key NATO members
plt.figure(figsize=(8, 8))
plt.pie(df['Defense_Spending_Billion_EUR'], labels=df['Country'], autopct='%1.1f%%', startangle=140)
plt.title('NATO Defense Spending Distribution (Billion EUR)')
plt.show()
```
